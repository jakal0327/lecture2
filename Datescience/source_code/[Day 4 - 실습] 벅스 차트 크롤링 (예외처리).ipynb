```python
from bs4 import BeautifulSoup
import pandas as pd
import urllib
from urllib.request import urlopen

from tqdm import tqdm_notebook

chart_date = []
chart_ranking = []
chart_title = []
chart_artist = []
chart_album = []

date = pd.date_range('2016-01-12', periods=20, freq='m')
```

```python
for today in tqdm_notebook(date):
    url = "https://music.bugs.co.kr/chart/track/day/total?chartdate={date}"
    html = url.format(date=urllib.parse.quote(today.strftime('%Y%m%d')))
    
    response = urlopen(html)
    soup = BeautifulSoup(response, 'html.parser')
    end = len( soup.find_all('p', 'title') )
    
    for n in range(end):
        artist = soup.find_all('p', 'artist')[n].a.string
        if artist == '아이유(IU)':
            chart_date.append(today)
            chart_ranking.append(soup.find_all('div', 'ranking')[n].strong.string)
            chart_title.append(soup.find_all('p', 'title')[n].a.string)
            chart_album.append(soup.find_all('a', 'album')[n+1].string)
```

```python
chart = pd.DataFrame({'date': chart_date,
                     'title': chart_title,
                     'ranking': chart_ranking })
chart
```

```python
chart['ranking'] = pd.to_numeric( chart['ranking'] )
chart_pivot = pd.pivot_table(chart, index='date', columns='title')
chart_pivot
```
